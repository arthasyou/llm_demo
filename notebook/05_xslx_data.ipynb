```python
import re

def remove_digits_from_start(text):
    pattern = r'^\d+ '  # 匹配以数字开头的模式
    result = re.sub(pattern, '', text)  # 删除匹配到的部分
    return result
```

```python
def replace_doctor(text):
    if text.startswith("医生"):
        result = "### 医生" + text[2:]
    else:
        result = text
    return result
```

```python
def replace_qa(text):
    if text.startswith("患者"):
        result = "<q>" + text[3:] + "</q>"
        return "q", result
    else:
        result = "<a>" + text[3:] + "</a>"
        return "a", result
```
