```python
def process_txt_file(file_path):
    # 打开文件并读取内容
    with open(file_path, 'r', ) as file:
        content = file.read()

    # 初始化结果列表
    data = []

    # 将文章分成每10个字符为一条数据
    for i in range(0, len(content), 1):
        data.append(content[i:i+2048])

    return data
```

```python

# 调用函数处理txt文件
file_path = "../data/txt/zyll1.txt"  # 替换为实际的文件路径
result = process_txt_file(file_path)
data = {'text': result}
```

```python
from datasets import Dataset
# 创建datasets
dataset = Dataset.from_dict(data)

# 保存为Hugging Face的datasets格式
dataset.save_to_disk("../data/datasets/zyll1")
```
